```
import re

def perform_operation(num1, op, num2):
    """Performs the specified arithmetic operation."""
    if op == '+':
        return num1 + num2
    elif op == '-':
        return num1 - num2
    elif op == '*':
        return num1 * num2
    elif op == '/':
        if num2 == 0:
            raise ValueError("Divisão por zero não permitida.")
        return num1 / num2
    else:
        raise ValueError(f"Operador '{op}' não suportado.")

print("--- Calculadora Python ---")
print("Digite a operação [ Ex: 10+5, 8 * 8, 100/2 ]")
print("Escreva 'sair' para encerrar o programa.\n")

while True:
    entrada = input("> ").strip().lower()

    if entrada == 'sair':
        print("\nEncerrando a calculadora... Até logo!")
        break

    # Use a regex to match num1, operator, and num2, allowing spaces around the operator
    match = re.match(r'(\d+\.?\d*)\s*([-+*/])\s*(\d+\.?\d*)', entrada)

    if not match:
        print("Erro: Formato inválido. Use algo como '10+5', '8 * 8', '100/2'.")
        continue

    try:
        num1_str, op, num2_str = match.groups()
        num1 = float(num1_str)
        num2 = float(num2_str)

        resultado = perform_operation(num1, op, num2)
        print(f"Resultado: {resultado:.2f}")
        print("-" * 20)

    except ValueError as e:
        print(f"Erro: {e}")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
```
